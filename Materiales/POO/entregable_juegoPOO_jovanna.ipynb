```python
#super clase guerrero

class Guerrero:
    def __init__(self, nombre):
        #se determina en base a cada tipo de guerrero
        self.nombre = nombre
        self.salud = 0  
        self.fuerza = 0 
        self.defensa = 0  
        self.velocidad = 0  


    def defender(self, daño):
        #reduce el daño basado en la defensa del guerrero
        daño_bloqueado = (daño - self.defensa)  #como hacer que el daño no sea negativo??? <-- REGRESAE!!
        self.salud -= daño_bloqueado
        print(f"{self.nombre} recibe {daño_bloqueado} puntos de daño. Salud restante: {self.salud}")

    def atacar(self, objetivo):
        #dar daño en base a la fuerza del guerrero
        print(f"{self.nombre} ataca a {objetivo.nombre}!")
        daño = self.fuerza
        objetivo.defender(daño)

    def sigue_vivo(self):
        #ver si el guerrero sigue vivo
        if self.salud > 0 :
            return (f'{self.nombre} sigue vivo! {(emoji.emojize(":grinning_face:"))}')
        elif self.salud == 0 :
            return (f'{self.nombre} ha muerto! {(emoji.emojize(":face_with_bags_under_eyes:"))}')

    def mostrar_estadisticas(self):
        #estadísticas del guerrero
        print(f"Guerrero: {self.nombre}")
        print(f"Salud: {self.salud}")
        print(f"Fuerza: {self.fuerza}")
        print(f"Defensa: {self.defensa}")
        print(f"Velocidad: {self.velocidad}")

    
class Caballero(Guerrero):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.salud = random.randint(100, 150)
        self.fuerza = random.randint(30, 40)
        self.defensa = random.randint(50, 70)
        self.velocidad = random.randint(10, 20)

    def atacar(self, daño): 
        #El Caballero realiza un ataque directo basado en su fuerza.
        daño = (self.fuerza * 1.2)
    def defender(self, defensa, daño): 
        #El Caballero bloquea un porcentaje del daño recibido, basado en su defensa.
        daño_recibido = daño - (defensa * 0.6)
```
