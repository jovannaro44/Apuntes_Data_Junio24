```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#url de la pagina a scrapear

url = 'https://books.toscrape.com/'
#contenido de la pagina
html = req.get(url).text 
```

```python
#sopita
soup = bs(html, 'html.parser') 
soup
```

```python
#entrar a la pagina del propio libro
#href
dominio = 'https://books.toscrape.com/'
url = dominio + 'index.html'

url
```

```python
urls = [dominio + link.attrs['href'] for link in soup.select('h3 a')]
urls
```

```python
#req que me haga otra sopa de lo que hay dentro del link del libro

for book_url in urls:
    book_response = req.get(book_url)
    book_soup = bs(book_response.text, 'html.parser')

book_soup
```

```python
print(book_soup.title.text)
```

```python
#extraigo los 20 libros que hay dentro de la pagina principal
#lista de links -- osea bucle de links
urls_books = [book_url for book_url in urls]
urls_books
```
